## Map CitiBike Data to CT level

### Load CitiBike station location data

```python
station_id=DataFrame.from_csv('Data/Station_id.csv')
#convert station_ids to string index
station_id.index=station_id.index.astype(str)
station_id.head()
```

### Create CitiBike station Geometry (Point) from Coordinates (Lat & Lon)

```python
import shapely
# combine lat and lon to one column
station_id['lonlat']=zip(station_id.lon,station_id.lat)
# Create Point Geometry for based on lonlat column
station_id['geometry']=station_id[['lonlat']].applymap(lambda x:shapely.geometry.Point(x))
station_id.head()
```

### Intersect the two geometries

```python
a=[]
# check for each citibike stations to see what CT they belong to
for i in range(len(station_id)):
    a.append(ct_shape[ct_shape.geometry.intersects(station_id.iloc[i].geometry)].BoroCT2010.values[0])
    print '\r',"%",str((i+1)*100./len(station_id))[:4],
station_id['BoroCT2010']=a
```

### Spatial Lag  (when neighboring observations affect one another)
Similarity of CT neighbors can be measured by means of spatial lag as follows:
$$BikeRideCountLag = \sum_j w_{i,j} BikeRideCount$$

```python
SummerMonthlyRide=ct_shape_SummerMonthlyAve_Zoom.SummerMonthlyAve
SummerMonthlyRideLag=ps.lag_spatial(qW_CT_summer, SummerMonthlyRide)
```

#### Create Quantiles for mapping

```python
# "when you don't know the nature of your data, the best way is quantile"

SummerMonthlyRideLagQ10 = ps.Quantiles(SummerMonthlyRideLag, k=10)
SummerMonthlyRideLagQ10
```

#### Map the summertime monthly rides lag

```python
f, ax = plt.subplots(figsize = (7,10))
ct_shape_SummerMonthlyAve_Zoom.assign(cl=SummerMonthlyRideLagQ10.yb).plot(column='cl', categorical=True, 
        k=10, cmap='hot_r', linewidth=0.1, ax=ax, 
        edgecolor='white', legend=True)
ax.set_axis_off()
plt.title("Summertime Monthly Ride Spatial Lag Deciles")
```

Fig 8: The tracts with highest spatial lag are all in on the west side and middle of lower and mid Manhattan. The farther you get from west-mid Manhattan, the lower the spatial lag.

## Global  Spatial Autocorrelation

```python
# what is happening here?
b,a = np.polyfit(SummerMonthlyRide, SummerMonthlyRideLag, 1)
b,a 
```

### Moran Scatterplot

```python
f, ax = plt.subplots(1, figsize=(9, 9))

plt.plot(SummerMonthlyRide, SummerMonthlyRideLag, '.', color='firebrick')

 # dashed vert at mean of the last year's PCI
plt.vlines(SummerMonthlyRide.mean(), SummerMonthlyRideLag.min(), SummerMonthlyRideLag.max(), linestyle='--')
 # dashed horizontal at mean of lagged PCI
plt.hlines(SummerMonthlyRideLag.mean(), SummerMonthlyRide.min(), SummerMonthlyRide.max(), linestyle='--')

# red line of best fit using global I as slope
plt.plot(SummerMonthlyRide, a + b*SummerMonthlyRide, 'r')
plt.title('Moran Scatterplot')
plt.ylabel('Spatial Lag of Summertime MonthlyRides')
plt.xlabel('Summertime Monthly Rides')
```

Fig 9: The more monthly rides there are in a census tract (in the summer), the more spatial lag those tracts have.

### Moran's I

```python
I_SummerMonthlyRide = ps.Moran(SummerMonthlyRide, qW_CT_summer)
```

## Local  Spatial Autocorrelation

```python
LMo_SummerMonthlyRide = ps.Moran_Local(SummerMonthlyRide, qW_CT_summer, permutations=999)
```

```python
LMo_SummerMonthlyRide.Is[0:5], LMo_SummerMonthlyRide.p_sim[0:5]
```

### LISA (Local Indicators of Spatial Autocorrelation)

```python
sigs = SummerMonthlyRide[LMo_SummerMonthlyRide.p_sim <= .001]
W_sigs = SummerMonthlyRideLag[LMo_SummerMonthlyRide.p_sim <= .001]
insigs = SummerMonthlyRide[LMo_SummerMonthlyRide.p_sim > .001]
W_insigs = SummerMonthlyRideLag[LMo_SummerMonthlyRide.p_sim > .001]
```

### Plot Significant LISA

```python
# why so small?
f, ax = plt.subplots(figsize=(18, 18))
plt.plot(sigs, W_sigs, '.', color='firebrick')
plt.plot(insigs, W_insigs, '.k', alpha=.2)
plt.text(s='$I = %.3f$' % I_SummerMonthlyRide.I, x=120000, y=5000, fontsize=18)

 # dashed vert at mean of the last year's PCI
plt.vlines(SummerMonthlyRide.mean(), SummerMonthlyRideLag.min(), SummerMonthlyRideLag.max(), linestyle='--')
 # dashed horizontal at mean of lagged PCI
plt.hlines(SummerMonthlyRideLag.mean(), SummerMonthlyRide.min(), SummerMonthlyRide.max(), linestyle='--')

# red line of best fit using global I as slope
plt.plot(SummerMonthlyRide, a + b*SummerMonthlyRide, 'r')
plt.title('Moran Scatterplot')
plt.ylabel('Spatial Lag of Monthly Rides in Summer')
plt.xlabel('Monthly Rides in Summer')
plt.title('Monthly Wummer Ridership Compared with Spatial Lag')
```

### Hot and Cold spots 

```python
sig_summer = LMo_SummerMonthlyRide.p_sim < 0.01
hotspots_summer = LMo_SummerMonthlyRide.q==1 * sig_summer
coldspots_summer = LMo_SummerMonthlyRide.q==3 * sig_summer
print "Number of Hotspots:", hotspots_summer.sum()
print "Number of Coldspots:", coldspots_summer.sum()
```

```python
from matplotlib import colors

hcmap = colors.ListedColormap(['grey','red','blue'])
hotcold_summer = hotspots_summer*1 + coldspots_summer*2

f, ax = plt.subplots(figsize=(7, 10))
ct_shape_SummerMonthlyAve_Zoom.assign(cl=hotcold_summer).plot(column='cl', categorical=True, 
        k=10, cmap=hcmap, linewidth=0.1, ax=ax, 
        edgecolor='white', legend=True)
ax.set_axis_off()
plt.title("Summertime Monthly Ave Spatial Lag Hotspots and Cold Spots")
```

Fig 13: I think this shows places with lots of spatial lag (hotspots) and places with little spatial lag (coldspots).
